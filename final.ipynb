#### Question 2.7 How are the years of experience related to the salaries?

```python
import matplotlib.pyplot as plt
dot_size= [ i/1000 for i in (df["Minimum Salary"])]
print(dot_size)
min_salary= (df["Minimum Salary"])
min_exp= (df["Minimum Year of Experience"])

plt.xlabel=("Minimum Salary")
plt.ylabel= ("Minimum Year of Experience")
             
plt.scatter(min_salary, min_exp, s = dot_size, c =min_salary)
plt.show()
```

As you increase in your years of experience, the likelihood for a higher salary also increased. This is seen with a weak but positive correlation of .203. Similarly, this linear relationship means that there is a correlation of less pay with fewer years of experience. This means, however, that due to a weak correlation, one cannot assume that having a higher level of experience will equate to higher pay. This indicates that salary levels off after a certain number of years in the workforce.  

#### Question 2.8 What are the common job duties? 

```python
import pandas
df = pandas.read_excel('Final Programming Project clean.xlsx')
for duty in df['Duties']:
    print (duty)
    
duties =''
for duty in df['Duties']:
    duties = duty + duties
print (duties)
import xlwt        

from collections import Counter        

from nltk.corpus import stopwords

stop = set(stopwords.words('english'))


book = xlwt.Workbook() 

sheet_test = book.add_sheet('duties') 

i = 0

sheet_test.write(i,0,'word') 

sheet_test.write(i,1,'count') 

sheet_test.write(i,2,'ratio') 

word_list = [i for i in duties.lower().split() if i not in stop]

word_total = word_list.__len__()

count_result =  Counter(word_list)

for result in count_result.most_common(20):

    i = i+1 

    sheet_test.write(i,0,result[0])

    sheet_test.write(i,1,result[1])

    sheet_test.write(i,2,(result[1]/word_total))


book.save('duties.xls')# define the location of your excel file
```

Common duties include research, analysis, identification of threats within data, among other intelligence expectations. Intelligence is the most commonly used word in relation to job duties, indicating that more administrative work is secondary. The implication of this visual is that the main duties of intelligence specialists are centered around finding information and understanding its implications.
